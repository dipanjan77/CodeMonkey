## Exploring and Analyzing data using Spark DataFrames (London Crime dataset)
#### Set data location and type

```python
import yaml

with open("config.yaml", 'r') as ymlfile:
    cfg = yaml.safe_load(ymlfile)

storage_account_name=cfg['storage_account_name']
storage_account_containername=cfg['storage_account_containername']
storage_account_accesskey=cfg['storage_account_accesskey']
file_location=cfg['file_location']
file_type="csv"

```

```python
spark.conf.set(
"fs.azure.account.key."+storage_account_name+".blob.core.windows.net",
  storage_account_accesskey)
```

### Read the data

```python
data=spark.read\
          .format(file_type)\
          .option("header","true")\
          .load(file_location)
```
